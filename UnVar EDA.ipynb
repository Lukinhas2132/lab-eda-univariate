#### Measures of Centrality:

```python
# Calculate the mean, median, and mode for the 'price' column
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]  # the first mode in case there are multiple

print(f"Mean Price:   {mean_price:.2f}")
print(f"Median Price: {median_price:.2f}")
print(f"Mode Price:   {mode_price:.2f}")
```

#### Measures of Dispersion:

```python
# Variance
variance_price = df['price'].var()
# Standard Deviation
std_price = df['price'].std()
# Range (max - min)
price_range = df['price'].max() - df['price'].min()
# Interquartile Range (IQR = Q3 - Q1)
q1 = np.percentile(df['price'], 25)
q3 = np.percentile(df['price'], 75)
iqr = q3 - q1

print(f"Variance of Price:    {variance_price:.2f}")
print(f"Standard Deviation:   {std_price:.2f}")
print(f"Price Range:          {price_range:.2f}")
print(f"Interquartile Range:  {iqr:.2f}")
```

# Unpacking Product Ratings

#### Measures of Centrality

```python
# Mean, Median, and Mode for 'stars'
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]  # first mode in case there are multiple

print(f"Mean Rating:   {mean_rating:.2f}")
print(f"Median Rating: {median_rating:.2f}")
print(f"Mode Rating:   {mode_rating:.2f}")
```

#### Measures of Dispersion

```python
# Variance, Standard Deviation, and Interquartile Range for 'stars'
variance_rating = df['stars'].var()
std_rating = df['stars'].std()

q1 = df['stars'].quantile(0.25)
q3 = df['stars'].quantile(0.75)
iqr_rating = q3 - q1

print(f"Variance of Rating:   {variance_rating:.2f}")
print(f"Standard Deviation:   {std_rating:.2f}")
print(f"Interquartile Range:  {iqr_rating:.2f}")
```

####  Shape of the Distribution.

```python
# Skewness and Kurtosis
# pandas has built-in methods for skewness and kurtosis:
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurtosis()

print(f"Skewness of Rating:  {skewness_rating:.2f}")
print(f"Kurtosis of Rating:  {kurtosis_rating:.2f}")
```

#### Visualizing the Rating Distribution

```python
plt.figure(figsize=(8, 5))
plt.hist(df['stars'], bins=10, edgecolor='black')
plt.xlabel("Star Rating")
plt.ylabel("Number of Products")
plt.title("Distribution of Product Ratings")
plt.show()
```
